## Задание 9:
Провести анализ пиковой активности студентов перед экзаменом (с использованием (Common Table Expression — CTE), оператор with).

Вывести, на какой неделе семестра студенты проявляли наибольшую активность в курсе в целом, и как эта активность распределяется между студентами-бюджетниками и контрактниками.

Пример вывода :

| name_osno | week_number	| avg_s_all	| avg_s_course_viewed |	avg_s_q_attempt_viewed |
|-----|------|------|------|------|
| бюджет |	14	| 125.45 |	45.67 |	32.12 |

```python
import psycopg2

conn = psycopg2.connect(
    dbname="my_db_Kashin",
    user="Kashin",
    password="Super_Strong_Pass_2026!_",
    host="localhost",
    port="5433"
)
cur = conn.cursor()

print("=" * 90)
print(" АНАЛИЗ ПИКОВОЙ АКТИВНОСТИ СТУДЕНТОВ ПЕРЕД ЭКЗАМЕНОМ (CTE)")
print("=" * 90)

#  Запрос с CTE (WITH)
query = """
WITH weekly_activity AS (
    -- CTE 1: Агрегируем активность по неделям и типу финансирования
    SELECT 
        name_osno,
        num_week AS week_number,
        AVG(s_all) AS avg_s_all,
        AVG(s_course_viewed) AS avg_s_course_viewed,
        AVG(s_q_attempt_viewed) AS avg_s_q_attempt_viewed,
        COUNT(*) as total_events
    FROM user_logs
    WHERE name_osno IS NOT NULL 
      AND num_week IS NOT NULL
    GROUP BY name_osno, num_week
),
peak_week AS (
    -- CTE 2: Находим неделю с максимальной общей активностью
    SELECT 
        week_number,
        SUM(total_events) as total_activity
    FROM weekly_activity
    GROUP BY week_number
    ORDER BY total_activity DESC
    LIMIT 1
)
-- Основной запрос: выводим активность по типу финансирования для пиковой недели
SELECT 
    wa.name_osno,
    wa.week_number,
    ROUND(wa.avg_s_all::NUMERIC, 2) AS avg_s_all,
    ROUND(wa.avg_s_course_viewed::NUMERIC, 2) AS avg_s_course_viewed,
    ROUND(wa.avg_s_q_attempt_viewed::NUMERIC, 2) AS avg_s_q_attempt_viewed
FROM weekly_activity wa
JOIN peak_week pw ON wa.week_number = pw.week_number
ORDER BY 
    CASE 
        WHEN wa.name_osno = 'бюджет' THEN 1
        WHEN wa.name_osno = 'контракт' THEN 2
        ELSE 3
    END;
"""

try:
    cur.execute(query)
    results = cur.fetchall()
    
    if results:
        # Вывод в формате таблицы как в примере
        print(f"\n{'name_osno':<15} | {'week_number':<11} | {'avg_s_all':<12} | {'avg_s_course_viewed':<20} | {'avg_s_q_attempt_viewed':<20}")
        print("-" * 90)
        
        for row in results:
            name_osno = row[0]
            week_number = row[1]
            avg_s_all = row[2]
            avg_s_course_viewed = row[3]
            avg_s_q_attempt_viewed = row[4]
            
            print(f"{name_osno:<15} | {week_number:<11} | {avg_s_all:<12} | {avg_s_course_viewed:<20} | {avg_s_q_attempt_viewed:<20}")
        
        print("\n Анализ завершен!")
        print(" CTE использованы:")
        print("   1. weekly_activity - агрегация активности по неделям и типу финансирования")
        print("   2. peak_week - определение недели с максимальной активностью")
    else:
        print(" Нет данных для анализа")
        
except Exception as e:
    print(f" Ошибка: {e}")
    print("\n Проверка доступных значений name_osno:")
    cur.execute("SELECT DISTINCT name_osno FROM user_logs WHERE name_osno IS NOT NULL LIMIT 10;")
    values = cur.fetchall()
    for val in values:
        print(f"   • '{val[0]}'")

finally:
    cur.close()
    conn.close()
```
